## Task 1 - Needleman-Wunsch
The code for Needleman-Wunsch contains three functions: alignment(string1, string2), check(char1, char2) and traverse_backwards(matrix, string1, string2). The alignment function creates a matrix that is populated with tuples of values and directions, the first value in the top leftmost corner being (0, 'none'). The check function checks if a substituion exists in the blosum50 matrix which is read into an array using NumPy. If there is a substitution, its value is returned else None is returned. The final function traverse_backwards moves through the matrix from the bottom rightmost corner and follows the directions stored with each value in a while loop to construct the strings for the alignment which are then reversed, as they are constructed backwards, and returned in a tuple. The cost value/d-value used in the algorithm is -8.

The result of the alignment of PAWHEAE and HEAGAWGHEE is('--P-AW-HEAE', 'HEAGAWGHE-E').

## Task 4 - HMM

In these functions a value is emitted and added to a string of emissions, provided it has not exceeded the number provided to them, and the state travelled to is added to an array of states. Each function uses Python's random.choice() method to choose whether to stay in the state or traverse to the other state as well as to choose a value to emit. Due to the very low probabilities of changing state, it is unlikely to see a change in states unless the number of emissions is very high.

## Task 5 - Viterbi Algorithm
The code has two functions: max_j(path1, transition1, path2, transition2, states) and viterbi(emissions, states, emission_probabilities_1, emission_probabilities_2, transitions). The function viterbi is the entry point into the application and it takes an array of values that have been emitted, an array of two states, a dictionary for the first state with keys that are the possible emission values and values that are the probabilities for these emissions, a dictionary of the same format but for the second state and an array of length four of transition probabilities in the order from state 1 to 1, 1 to 2, 2 to 2 and 2 to 1. The Viterbi algorithm will return the most probable path through the state machine given the array of emissions and display a line graph of these states across time with x-axis labels at the start, finish and any points where the current state differs from that in the previous timestep. The start state is randomly chosen with equal probability. The max_j function is a helper function to calculate the maximum of all states in the previous timestep multiplied by the transition probability from that state.
